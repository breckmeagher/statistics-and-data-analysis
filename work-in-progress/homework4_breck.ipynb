### **Problem 1: Generating Data**

Generate two 2D Gaussian clusters:
$$
x0 ∼ N ((0, 0), I) \\
x1 ∼ N ((µ, 0), I)
$$

where I is the identity matrix.  
1. Plot the data for µ = 3
2. Label each point by its class

---

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(25)
n_samples = 100
I = np.eye(2)

# generate clusters
x0 = np.random.multivariate_normal([0, 0], I, n_samples)
x1 = np.random.multivariate_normal([3, 0], I, n_samples)

# plotting
plt.figure(figsize=(8, 6))
plt.scatter(x0[:, 0], x0[:, 1], c='blue', label=r'Class 0: $N((0,0), I)$')
plt.scatter(x1[:, 0], x1[:, 1], c='orange', label=r'Class 1: $N((3,0), I)$')
plt.title(r'2D Gaussian Clusters ($ = 3$)')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```
